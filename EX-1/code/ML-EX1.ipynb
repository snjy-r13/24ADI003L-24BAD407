```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(
    "/kaggle/input/ecommerce-data/data.csv",
    encoding="ISO-8859-1"
)

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())

df_clean = df.dropna(subset=["Description"]).copy()

df_clean["Sales"] = df_clean["Quantity"] * df_clean["UnitPrice"]

sales_per_product = (
    df_clean
    .groupby("Description")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

top_10_products = sales_per_product.head(10)

plt.figure(figsize=(12,6))
top_10_products.plot(kind="bar")
plt.title("Sales per Product (Top 10)")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.show()

plt.figure(figsize=(12,6))
plt.plot(top_10_products.index, top_10_products.values, marker="o")
plt.title("Sales per Product (Top 10)")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=75)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())

columns_with_zero_as_missing = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[columns_with_zero_as_missing] = df[columns_with_zero_as_missing].replace(0, np.nan)

print(df.isnull().sum())

plt.figure(figsize=(10,6))
plt.hist(df["Glucose"].dropna(), bins=30)
plt.title("Glucose Level Distribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(df["Glucose"].dropna(), vert=False)
plt.title("Glucose Level Boxplot")
plt.xlabel("Glucose")
plt.show()

plt.figure(figsize=(10,6))
plt.hist(df["Age"], bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(df["Age"], vert=False)
plt.title("Age Boxplot")
plt.xlabel("Age")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/housing-price-prediction/Housing.csv")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())

plt.figure(figsize=(8,6))
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(df["bedrooms"], df["price"])
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms vs Price")
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(df["bathrooms"], df["price"])
plt.xlabel("Bathrooms")
plt.ylabel("Price")
plt.title("Bathrooms vs Price")
plt.show()

plt.figure(figsize=(10,8))
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/bank-customer-churn-dataset/Bank Customer Churn Prediction.csv")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.columns)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

x_col = numeric_cols[0]
y_col = numeric_cols[1]

plt.figure(figsize=(10,6))
plt.hist(df[x_col], bins=30)
plt.xlabel(x_col)
plt.ylabel("Frequency")
plt.title(f"{x_col} Distribution")
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(df[x_col], vert=False)
plt.xlabel(x_col)
plt.title(f"{x_col} Boxplot")
plt.show()

plt.figure(figsize=(10,6))
plt.hist(df[y_col], bins=30)
plt.xlabel(y_col)
plt.ylabel("Frequency")
plt.title(f"{y_col} Distribution")
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(df[y_col], vert=False)
plt.xlabel(y_col)
plt.title(f"{y_col} Boxplot")
plt.show()

plt.figure(figsize=(10,6))
plt.scatter(df[x_col], df[y_col])
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title(f"{x_col} vs {y_col}")
plt.show()

plt.figure(figsize=(10,8))
corr_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
